# Students t test

Student's t-test:
Tests whether the means of two independent samples are significantly different.
One-sample student's t-test
Test a sample with a known standard value.

**Assumptions**

- Observations in sample are independent and identically distributed.
- Observations in sample are normally distributed.
 
 **Interpretation**

**H0:** the means of the sample is equal to the known value.

**H1:** the means of the sample is unequal to the known value.

# 2 **Independent student's t-test**

**Assumptions**

- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.

**Interpretation**

**H0:** the means of the samples are equal.

**H1:** the means of the samples are unequal.

# 3 **Paired student's t-test**

Tests whether the means of two paired samples are significantly different.
**Assumptions**

- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.
- Observations across each sample are paired.
**Interpretation**

**H0:** the means of the samples are equal.

**H1:** the means of the samples are unequal.

```python
# Check the age and compare with a known value of 45years

ttest_1samp(df1["age"],45)
stat, p = ttest_1samp(df1["age"],45)

print("stat=%.3f,p=%.3f"%(stat,p))

# adding conditioal argument for ease
if p > 0.05:
    print("Probably the same distribution")
else:
    print("Probably different distributions")
```

## *TWo Sample ttest* ##
Independent student's t-test

**Assumptions**

- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.

**Interpretation**

**H0:** the means of the samples are equal.

**H1:** the means of the samples are unequal.

```python
# We will compare age and fare of male vs female passengers

# Splitting datasets
df_male=df1.loc[df1["sex"]=="male"]
df_female=df1.loc[df1["sex"]=="female"]
```

```python
# Import libraries

from scipy.stats import ttest_ind
stat,p=ttest_ind(df_male["fare"],df_female["fare"])

print("stat=%.3f,p=%.3f" % (stat,p))

# adding conditional argument for ease
if p >0.05:
    print("Probably the same distribution")
else:
    print("Probably different distributions")
```

**Paired student's t-test**

Tests whether the means of two paired samples are significantly different.
**Assumptions**

- Observations in each sample are independent and identically distributed.
- Observations in each sample are normally distributed.
- Observations in each sample have the same variance.
- Observations across each sample are paired.
**Interpretation**

**H0:** the means of the samples are equal.

**H1:** the means of the samples are unequal.

```python
# select only two classes
df_male_first=df_male.loc[df_male["class"]=="First"]
df_male_second=df_male.loc[df_male["class"]=="Second"]
df_male_third=df_male.loc[df_male["class"]=="Third"]
```

```python
df_1st=df_male_first.sample(n=100)
df_3rd=df_male_third.sample(n=100)

print("The number of instances in 1st class are ",df_1st.shape)
print("The number of instances in 3rd class are ",df_3rd.shape)
```
